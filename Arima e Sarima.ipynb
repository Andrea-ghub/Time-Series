# Arima

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA

# Caricare il dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Visualizzare i primi dati
print(df.head())

# Plot della serie temporale
plt.figure(figsize=(10,5))
plt.plot(df, label="Numero di passeggeri")
plt.xlabel("Anno")
plt.ylabel("Numero di passeggeri")
plt.title("Serie temporale - AirPassengers")
plt.legend()
plt.show()

```

```python
# Test di Dickey-Fuller per verificare la stazionarietà
def test_stationarity(timeseries):
    result = adfuller(timeseries)
    print(f'Test di Dickey-Fuller:')
    print(f'Statistica: {result[0]:.4f}')
    print(f'P-value: {result[1]:.4f}')
    print(f'Valori critici: {result[4]}')

test_stationarity(df['Passengers'])
```

```python
# Differenziazione per rendere la serie stazionaria
df['Passengers_diff'] = df['Passengers'].diff()
df.dropna(inplace=True)

# Riprova il test di stazionarietà dopo la differenziazione
test_stationarity(df['Passengers_diff'])
```

```python
# Plot della serie differenziata
plt.figure(figsize=(10,5))
plt.plot(df['Passengers_diff'], label="Serie Differenziata")
plt.xlabel("Anno")
plt.ylabel("Differenza Numero di passeggeri")
plt.title("Serie temporale differenziata")
plt.legend()
plt.show()
```

```python
# Creazione del modello ARIMA (p=2, d=1, q=2 scelti empiricamente)
model = ARIMA(df['Passengers'], order=(2,1,2))
model_fit = model.fit()

# Output del sommario del modello
print(model_fit.summary())
```

```python
# Previsione per i prossimi 12 mesi
forecast = model_fit.forecast(steps=12)
forecast_index = pd.date_range(df.index[-1], periods=12, freq='M')

# Plot dei risultati
plt.figure(figsize=(10,5))
plt.plot(df['Passengers'], label="Dati reali")
plt.plot(forecast_index, forecast, label="Previsione", linestyle="dashed", color="red")
plt.xlabel("Anno")
plt.ylabel("Numero di passeggeri")
plt.title("Previsione con ARIMA")
plt.legend()
plt.show()
```

# Sarima

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# Plot della serie temporale
plt.figure(figsize=(10,5))
plt.plot(df, label="Numero di passeggeri")
plt.xlabel("Anno")
plt.ylabel("Numero di passeggeri")
plt.title("Serie Temporale - AirPassengers")
plt.legend()
plt.show()
```

```python
test_stationarity(df['Passengers'])
```

```python
# Definizione del modello SARIMA (p,d,q) x (P,D,Q,s)
# (p,d,q) =  (1,1,1)  -> Parte non stagionale
# (P,D,Q,s) = (1,1,1,12)  -> Parte stagionale con periodicità di 12 mesi

model = SARIMAX(df['Passengers'], 
                order=(1,1,1), 
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False, 
                enforce_invertibility=False)

model_fit = model.fit()

# Output del sommario del modello
print(model_fit.summary())
```

```python
# Previsione per i prossimi 12 mesi
forecast = model_fit.forecast(steps=24)
forecast_index = pd.date_range(df.index[-1], periods=24, freq='M')

# Plot dei risultati
plt.figure(figsize=(10,5))
plt.plot(df['Passengers'], label="Dati reali")
plt.plot(forecast_index, forecast, label="Previsione SARIMA", linestyle="dashed", color="red")
plt.xlabel("Anno")
plt.ylabel("Numero di passeggeri")
plt.title("Previsione con SARIMA")
plt.legend()
plt.show()
```
